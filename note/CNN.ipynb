```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests

def load_data():
    url = 'http://43.202.66.118:8016/all'
    r = requests.get(url).json()
    return r

data = load_data()
df = pd.DataFrame(data)
df
```

```python
df['request_time'] = pd.to_datetime(df['request_time'])
df['request_time'] = df['request_time'].dt.strftime('%Y-%m-%d %H')

df['prediction_time'] = pd.to_datetime(df['prediction_time'])
df['prediction_time'] = df['prediction_time'].dt.strftime('%Y-%m-%d %H')

r_time = df.groupby('request_time').size()
p_time = df.groupby('prediction_time').size()
```

```python
plt.title('Requests by Date and Time')
plt.bar(r_time.index, r_time.values)
plt.plot(p_time.index, p_time.values, marker='o', color='r')
plt.xlabel('Date and Time')  # x축 레이블
plt.ylabel('Count')  # y축 레이블
plt.xticks(rotation = 45)
plt.show()
```

```python
user = df.groupby('request_user').size()

plt.title('Requests by User')
plt.bar(user.index, user.values)
plt.plot(user.index, user.values, marker='o', color='r')
plt.xlabel('Request User')  # x축 레이블
plt.ylabel('Count')  # y축 레이블
plt.xticks(rotation = 45)
plt.show()
```

```python
index = np.arange(2)
bar_width = 0.5

model = df.groupby('prediction_model').size()

plt.title('Requests by User')
plt.bar(index, user.index, user.values, color='b')
plt.bar(index+bar_width, model.index, model.values, color='r')
plt.xlabel('prediction_model')  # x축 레이블
plt.ylabel('Count')  # y축 레이블
plt.xticks(rotation = 90)
plt.show()
```
